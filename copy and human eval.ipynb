```python
import re

def read_txt(directory):
    
    # read data as list of CLEANED LOWER-CASE sentences, separated by ##
    
    file = open(directory, 'r', encoding='utf-8')
    file_data = file.read() 
    all_data = file_data.split('\n')

    while "" in all_data:
        all_data.remove("")
    return all_data
```

```python
# debug
N = 1000

rewritten = read_txt('unseen_from_bert-base-uncased_predicted_by_bert-base-uncased_rewritten_data.txt')[:N]
original = read_txt('all_data.txt')[:N]
generated = read_txt('bert_rewrite_blenderbot_small-90M_generate.txt')[:N]

with open('copy_test.txt', 'w') as f:
    f.write('')
    
with open('copy_test.txt', 'a') as f:
    for o, r, g in tqdm(zip(original, rewritten, generated)):
        if o != r:
            for o_word, r_word in zip(o.split(), r.split()):
                if o_word != r_word and not re.match(r"\W", r_word):
                    g = re.sub('\\b'+r_word+'\\b', o_word, g)
        f.write(g)
        f.write('\n')
    f.close()
```
